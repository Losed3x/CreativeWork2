Маша и Катя делают домашнее задание по Обществознанию, в котором нужно Написать Блага и распределить их по группам. Но учитель боится, что ученики могут у друг-друга списать. Помогите ему сделать программу, при которой если обнаружено в двух множествах повторяющиеся (одно или два), выводит "Под рассмотрением", если повторяющихся три или больше, программа выводит "Работы списаны". Если повторений нет, программа выводит "работы не списаны".

---


 |Формат Ввода:
1. Список Благ Маши через пробел
2. Список Благ Кати через пробел

---


 |Формат Вывода:

Одна из фраз, описанных в условии, смотря на количество повторений.


---




 |Примеры:

 Ввод:

Вода учеба кровать

Вода жилье пианино

Вывод:

-
Под рассмотрением

---
Ввод:

 Бумага Тетрадь Карандаш телефон

 компьютер Бумага Карандаш Тетрадь


Вывод:

Работы Списаны



```python
M = set(input())
K = set(input())
a = M&K
a.remove(" ")
if len(a)==0:
  print("работы не списаны")
if len(a)==1 or len(a)==2:
  print("Под рассмотрением")
if len(a) >= 3:
  print("Работы списаны")
```

На пляж группой из 10 человек уехали ребята. Оказалось, что у некоторых из них одинаковые фамилии. Сделайте программу, выводящую повторяющиеся Имя и ее количество повторений.

---
Формат ввода:
10 имен, каждая с новой строчки.

---


Формат вывода:
Имя - ее количество повторений ( Выводить ТОЛЬКО имена, повторяющиеся 2 раза или больше)

---

Пример:

Ввод:


Катя
Женя
Катя
Соня
Иван
Илья
Катя
Иван
Саша
Коля

Вывод:

Катя - 3
Иван - 2


```python
a = dict()
n = int(input())
for i in range(n):
   name = input().strip()
    if name in a:
        a[name] += 1
    else:
        a[name] = 1
if a[name]==1:
      a.pop(name)
print(a)
```
